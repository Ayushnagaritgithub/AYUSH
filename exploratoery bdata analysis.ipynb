what is the location with faster roller coaster(minimun is 10)?

```python
df['location'].value_counts()
```

```python
ax=df.query('location!=other"').groupby('location')['Speed_mph']
.agg(['mean','count'])
.query('count>=10')
.sort_values('mean')
.plot(Kind='bar')
ax.set.Xlabel('Average Coaster Speed')
plt.show()
```
